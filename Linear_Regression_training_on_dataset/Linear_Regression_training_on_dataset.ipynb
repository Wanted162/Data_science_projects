```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
```

```python
df = pd.read_csv(url)
print(df.head())
```

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['horsepower'] = df1['horsepower'].fillna(0)
```

```python
y = df1['mpg']     ### dependent variable
X = df1[['cylinders', 'displacement', 'horsepower', 'weight',
       'acceleration']]  ##### independent variable
```

```python
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.80, random_state=1234)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_pred_train = lr.predict(X_train)
```

```python
print("the training accuracy is:")
r2_score(y_train, X_pred_train)
```

```python
X_pred_test = lr.predict(X_test)
```

```python
print("the testing accuracy is:")
r2_score(y_test, X_pred_test)
```

```python
ridge = Ridge()
```

```python
ridge.fit(X_train, y_train)
```

```python
X_ridge_test_pred = ridge.predict(X_test)
X_ridge_test_pred
```

```python
print("the ridge testing accuracy is:")
r2_score(y_test, X_ridge_test_pred)
```

```python
X_lasso_train_pred = lasso.predict(X_train)
X_lasso_train_pred
```

```python
print("the lasso training accuracy is:")
r2_score(y_train, X_lasso_train_pred)
```

```python
lasso.fit(X_train, y_train)
```

```python
X_lasso_test_pred = lasso.predict(X_test)
X_lasso_test_pred
```

```python
print("the lasso testing accuracy is:")
r2_score(y_test, X_lasso_test_pred)
```
