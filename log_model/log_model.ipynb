```python
import pandas as pd
import time
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
##### packages for model diagnostics
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
```

```python
%%time

#### importind data
def read_data():
  path = 'https://raw.githubusercontent.com/Wanted162/Data_science_projects/main/log_model/Students.csv'
  data = pd.read_csv(path)
  return data
```

```python
%%time
## calling the read func
data = read_data()
print(data)
```

```python
data.STATUS.value_counts()
```

```python
def data_prep(data, t):
    X = data[['JEEPRE', 'JEEMAINS']]
    y = data['STATUS']
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=t, random_state=365, stratify=y)   #### when doing classification use stratify
    return X_train, X_test, y_train, y_test
```

```python
### calling the data prep func
%%time
X_train, X_test, y_train, y_test = data_prep(data, 0.3)
```

```python
def log_model(X_train, y_train, X_test):
  lg_model = LogisticRegression()
  lg_model_fit = lg_model.fit(X_train, y_train)
  pred = lg_model_fit.predict(X_test)
  return pred

```

```python
%%time
pred = log_model(X_train, y_train, X_test)
pred
```

```python
def model_log_diag(y_test,pred):
  acc=accuracy_score(y_test, pred)
  prc=precision_score(y_test,pred)
  clf = classification_report(y_test,pred)
  conf=confusion_matrix(y_test,pred)
  return acc,prc,clf,conf
```

```python
acc,prc,clf,conf = model_log_diag(y_test,pred)
print("The accuracy score is - ",acc)
print("The precision score is - ",prc)
print("The classification report is - ",clf)
print("The confusion matrix is - ",conf)
```

```python
def function_call():
  data = read_data()
  X_train, X_test, y_train, y_test = data_prep(data)
  pred = log_model(X_train, y_train, X_test)
  return
```

```python
def ds_tree(X_train,y_train,X_test):
  model_clf = DecisionTreeClassifier(criterion='entropy', max_depth=5)
  train_model = model_clf.fit(X_train,y_train)
  predd = train_model.predict(X_test)
  return predd
```

```python
accuracy_score(y_test, predd)
```

```python
def rndf_tree(X_train,y_train,X_test):
  model_rdf = RandomForestClassifier(criterion='entropy', max_depth=5)
  train_model = model_rdf.fit(X_train,y_train)
  pred1 = train_model.predict(X_test)
  return pred1
```

```python
pred1 = rndf_tree(X_train,y_train,X_test)
```

```python
accuracy_score(y_test, pred1)
```
