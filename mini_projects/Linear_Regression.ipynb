```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'id':[1,2,3,4,5,6,7], 'name':['A','B','C','D','E','F','G']})
print(df1)
```

```python
df2 = pd.DataFrame({'id':[3,5,6], 'marks':[200,256,180]})
print(df2)
```

```python
print("innner join")
df3 = df1.merge(df2, on='id', how='inner')
print(df3)
```

```python
print("left join")
df4 = df1.merge(df2, on='id', how='left')
print(df4)
```

```python
print("right join")
df5 = df1.merge(df2, on='id', how='right')
print(df5)
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
```

```python
data = pd.read_csv(url)
print(data.info())
```

```python
data1 = data[['mpg', 'weight']]
data1
```

```python
x = data1['weight'].values
x = x.reshape(-1,1)  ##### It applie only to simple regression model
print(x)
```

```python
y = data1['mpg'].values
y = y.reshape(-1,1)  ##### It applie only to simple regression model
print(y)
```

```python
#### importing scikit learn for algorithmns
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
print(model)
```

```python
%%time
#### trianing the Model/system - By fitting my data
model_fit = model.fit(x,y)
print(model_fit)
```

```python
model_fit.coef_
```

```python
model_fit.intercept_
```

```python
predict = model_fit.predict(x)  # will be done on test data for validation and after that raw data
predict
```
