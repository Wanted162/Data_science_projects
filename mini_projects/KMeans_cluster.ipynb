```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/DataScience75/Batch-88/main/iris.json"
```

```python
data = pd.read_json(url)
print(data.head(2))
```

```python
X = data.iloc[:,0:4]
X
```

```python
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
X_scaled
```

```python
km = KMeans(n_clusters=5)
km.fit(X_scaled)
pred = km.predict(X_scaled)
pred
```

```python
km.cluster_centers_
```

```python
customer_reviews=[]
customer_reviews.append("Great product!Rahul Have a try")
customer_reviews.append("Customer service was excellent")
customer_reviews.append("Not satisfy with the quality needed")
```

```python
print(customer_reviews)
```

```python
print("customer reviews")
for review in customer_reviews:
  print("- " + review)
```

```python
rahul_review=[]
patern_to_find = r'Rahul?'  #### ? denotes after rahul if there is any patern bring those sentences also
```

```python
for s in customer_reviews:
  if re.search(patern_to_find,s):
    rahul_review.append(s)
```

```python
print(rahul_review)
```

```python
g_review=[]
pattern_to_find = r'^G'  #### ^ start of the string with letter G
```

```python
for g in customer_reviews:
  if re.search(pattern_to_find,g):
    g_review.append(g)
```

```python
print(g_review)
```
