```python
import pandas as pd
import numpy as np
import sklearn.linear_model
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/content/Salary_Data.csv")
```

```python
# prompt: using df1 make a prediction model for salary hike using linear regression model

import matplotlib.pyplot as plt
# Split the data into training and testing sets
X = df1.iloc[:, :-1].values
y = df1.iloc[:, 1].values
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=0)

# Fit the linear regression model
regressor = sklearn.linear_model.LinearRegression()
regressor.fit(X_train, y_train)

# Predict the salary hike
y_pred = regressor.predict(X_test)

# Visualize the results
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# prompt: show the accuracy of the df1 model

from sklearn.metrics import r2_score
acc = r2_score(y_test, y_pred)
print("accuracy:", acc)
```

```python
df2 = pd.read_csv("/content/delivery_time.csv")
```

```python
# prompt: using df2 predict the delivery time using sorting time using Linear regression model

import matplotlib.pyplot as plt
# Extract the relevant columns
X = df2[['Sorting Time']].values
y = df2['Delivery Time'].values

# Split the data into training and testing sets
X_train1, X_test1, y_train1, y_test1 = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=0)

# Fit the linear regression model
regressor = sklearn.linear_model.LinearRegression()
regressor.fit(X_train1, y_train1)

# Predict the delivery time
y_pred1 = regressor.predict(X_test1)

# Visualize the results
plt.scatter(X_train1, y_train1, color='red')
plt.plot(X_train1, regressor.predict(X_train1), color='blue')
plt.title('Delivery Time vs Sorting Time (Training set)')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()

plt.scatter(X_test1, y_test1, color='red')
plt.plot(X_train1, regressor.predict(X_train1), color='blue')
plt.title('Delivery Time vs Sorting Time (Test set)')
plt.xlabel('Sorting Time')
plt.ylabel('Delivery Time')
plt.show()
```

```python
# prompt: show the accuracy of the df2 model

accuracy = r2_score(y_test1, y_pred1)
print('Accuracy:', accuracy)
```
